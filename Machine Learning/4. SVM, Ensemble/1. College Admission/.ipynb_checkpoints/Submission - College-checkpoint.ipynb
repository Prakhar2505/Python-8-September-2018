```python
import pandas as pd
```

```python
data = pd.read_csv('College.csv')
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
college_label = LabelEncoder()
```

```python
data['Private'] = college_label.fit_transform(data['Private'])
```

```python
college_label.classes_
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
training_data = data.copy()
```

```python
output_data = data['Private'].copy()
```

```python
training_data.columns[1:]
```

```python
training_data = training_data[training_data.columns[1:]]
```

```python
training_data
```

```python
X, x_test, Y, y_test = train_test_split(training_data, output_data, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X, Y)
```

```python
predictions = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
accuracy_score(y_test, predictions)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize=(18, 8))
sns.heatmap(data.corr(), annot=True)
```

## Changing the data using Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scale_data = training_data.copy()
```

```python
scale_data = scaler.fit_transform(scale_data)
```

```python
scale_data = pd.DataFrame(scale_data)
```

```python
scale_data.columns = training_data.columns
```

```python
scale_data
```

## Making predictions again

```python
X, x_test, Y, y_test = train_test_split(scale_data, output_data, test_size=0.3, random_state=42)
```

```python
model.fit(X, Y)
```

```python
predictions = model.predict(x_test)
```

```python
accuracy_score(y_test, predictions)
```

## Using Grid Search to find the best hyper parameter

```python
parameters = {
    'kernel': ('linear', 'rbf'),
    'C': [0.01, 0.1, 1, 10, 100]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(model, parameters, cv=5)
```

```python
grid.fit(X, Y)
```

```python
predictions = grid.predict(x_test)
```

```python
accuracy_score(y_test, predictions)
```
